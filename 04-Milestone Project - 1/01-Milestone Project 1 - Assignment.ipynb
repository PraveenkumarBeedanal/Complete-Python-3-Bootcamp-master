```python
def intro_game():
    print('Welcome to Tic Tac Toe')
    player1_choice = input('Player 1: Do you want to be X or O?: ')
    print('Player 1 will go first')

    status = input('Are you ready to play Yes or No: ')

    return player1_choice, status
```

```python
def match_1_block_forward_diff(i: int, input_list: list) -> bool:
    if input_list[i] == 'X' and input_list[i+1] == 'X' and input_list[i+2] == 'X':
        print('Player1 is won')
        return True
    elif input_list[i] == 'O' and input_list[i+1] == 'O' and input_list[i+2] == 'O':
        print("Player2 is won")
        return True
```

```python
def match_2_block_forward_diff(i: int, input_list: list) -> bool:
    if input_list[i] == 'X' and input_list[i+2] == 'X' and input_list[i+4] == 'X':
        print('Player1 is won')
        return True
    elif input_list[i] == 'O' and input_list[i+2] == 'O' and input_list[i+4] == 'O':
        print("Player2 is won")
        return True
```

```python
def match_3_block_forward_diff(i: int, input_list: list) -> bool:
    if input_list[i] == 'X' and input_list[i+3] == 'X' and input_list[i+6] == 'X':
        print('Player1 is won')
        return True
    elif input_list[i] == 'O' and input_list[i+3] == 'O' and input_list[i+6] == 'O':
        print("Player2 is won")
        return True
```

```python
def match_4_block_forward_diff(i: int, input_list: list) -> bool:
    if input_list[i] == 'X' and input_list[i+4] == 'X' and input_list[i+8] == 'X':
        print('Player1 is won')
        return True
    elif input_list[i] == 'O' and input_list[i+4] == 'O' and input_list[i+8] == 'O':
        print("Player2 is won")
        return True
```

```python
def match_1_block_forward_diff_opposite_direction(i: int, input_list: list) -> bool:
    if input_list[i] == 'X' and input_list[i-1] == 'X' and input_list[i-2] == 'X':
        print('Player1 is won')
        return True
    elif input_list[i] == 'O' and input_list[i-1] == 'O' and input_list[i-2] == 'O':
        print("Player2 is won")
        return True
```

```python
def match_2_block_forward_diff_opposite_direction(i: int, input_list: list) -> bool:
    if input_list[i] == 'X' and input_list[i-2] == 'X' and input_list[i-4] == 'X':
        print('Player1 is won')
        return True
    elif input_list[i] == 'O' and input_list[i-2] == 'O' and input_list[i-4] == 'O':
        print("Player2 is won")
        return True
```

```python
def match_3_block_forward_diff_opposite_direction(i: int, input_list: list) -> bool:
    if input_list[i] == 'X' and input_list[i-3] == 'X' and input_list[i-6] == 'X':
        print('Player1 is won')
        return True
    elif input_list[i] == 'O' and input_list[i-3] == 'O' and input_list[i-6] == 'O':
        print("Player2 is won")
        return True
```

```python
def match_4_block_forward_diff_opposite_direction(i: int, input_list: list) -> bool:
    if input_list[i] == 'X' and input_list[i-4] == 'X' and input_list[i-8] == 'X':
        print('Player1 is won')
        return True
    elif input_list[i] == 'O' and input_list[i-4] == 'O' and input_list[i-8] == 'O':
        print("Player2 is won")
        return True
```

```python
def full_game_flow():
    
    # get player1 mark choice and whether he is ready or not
    player1_choice, status = intro_game()

    choices = ['X', 'O']

    choices.remove(player1_choice)

    player2_choice = choices[0]

    if status.lower() == 'no':
        exit        

    # display the board and take user input untill someone won

    won = False
    count = 1

    input_list = [' ', ' ', ' ', ' ', ' ', ' ', ' ', ' ', ' '] 

    ## print the empty box for once
    print('\n')
    print('   |    |   ')
    print(f' {input_list[0]} |  {input_list[1]} | {input_list[2]} ')
    print('   |    |   ')
    print('-----------')
    print('   |    |   ')
    print(f' {input_list[3]} |  {input_list[4]} | {input_list[5]} ')
    print('   |    |   ')
    print('-----------')
    print('   |    |   ')
    print(f' {input_list[6]} |  {input_list[7]} | {input_list[8]} ')
    print('   |    |   ')


    while not won:
        
        # take the position from the user
        user_index = int(input('Choose your next position: (1-9)'))  

        # 1st input is always from player1, hence odd inputs are from player1 and even inputs are from player2
        if count % 2 != 0:
            input_list[user_index-1] = player1_choice
            print(input_list)
        else:
            input_list[user_index-1] = player2_choice
            print(input_list)

        # show the box with user input
        print('\n')
        print('   |    |   ')
        print(f' {input_list[0]} |  {input_list[1]} | {input_list[2]} ')
        print('   |    |   ')
        print('-----------')
        print('   |    |   ')
        print(f' {input_list[3]} |  {input_list[4]} | {input_list[5]} ')
        print('   |    |   ')
        print('-----------')
        print('   |    |   ')
        print(f' {input_list[6]} |  {input_list[7]} | {input_list[8]} ')
        print('   |    |   ')

        ## check win condition before saying tied
        for i in range(len(input_list)):
            if i == 0:
                ## check for the 0, 1, 2 condition
                won = match_1_block_forward_diff(i, input_list)
                if won: break
                ## check for 0, 3, 6 conditon
                won = match_3_block_forward_diff(i, input_list)
                if won: break
                ## check for 0, 4, 8 condition
                won = match_4_block_forward_diff(i, input_list) 
                if won: break
            elif i == 1:
                ## check for 1, 4, 7 condition
                won = match_3_block_forward_diff(i, input_list)
                if won: break
            elif i == 2:
                ## check for 2, 5, 8 condition
                won = match_3_block_forward_diff(i, input_list)
                if won: break
                ## check for 2, 4, 6 condition
                won = match_2_block_forward_diff(i, input_list)
                if won: break
                ## check for 2, 1, 0 condtion
                won = match_1_block_forward_diff_opposite_direction(i, input_list)
                if won: break
            elif i == 3:
                ## check for 3, 4, 5 condition
                won = match_1_block_forward_diff(i, input_list)
                if won: break
            elif i == 5:
                ## check for 5, 4, 3 condition
                won = match_1_block_forward_diff_opposite_direction(i, input_list)
                if won: break
            elif i == 6:
                ## check for 6, 7, 8 condition
                won = match_1_block_forward_diff(i, input_list)
                if won: break
                ## check for 6, 4, 2 condition
                won = match_2_block_forward_diff_opposite_direction(i, input_list)
                if won: break
                ## check for 6, 3, 0 condition
                won = match_3_block_forward_diff_opposite_direction(i, input_list)
                if won: break
            elif i == 7:
                ## check for 7, 4, 1 condition
                won = match_3_block_forward_diff_opposite_direction(i, input_list)
                if won: break
            elif i == 8:
                ## check for 8, 4, 0
                won = match_4_block_forward_diff_opposite_direction(i, input_list)
                if won: break
                ## check for 8, 7, 6
                won = match_1_block_forward_diff_opposite_direction(i, input_list)
                if won: break
                ## check for 8, 5, 2 condition
                won = match_3_block_forward_diff_opposite_direction(i, input_list)
                if won: break

        # user can input upto 10 time only because there are only 9 boxes
        # even after 9th time, there is no result. Show it as tied
        if count == 10:
            print('Game tied')
            break

        count += 1
```

```python
full_game_flow()
```
